```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import copy
import pickle
```

```python
input_csv = '/home/acarlier/code/OrnithoScope/data/csv/input.csv'
task_csv = '/home/acarlier/code/OrnithoScope/data/csv/Ornithotasks - CVAT_task.csv'
pickle_saving_path = '/home/acarlier/code/OrnithoScope/data/pickles'
columns_names_csv = ["split_value", "file_path", "label",
                      "x_min", "y_min", "empty_1", "empty_2",
                      "x_max", "y_max", "empty_3"]
df_input = pd.read_csv(input_csv, names = columns_names_csv)
df_task = pd.read_csv(task_csv,header = 0)
```

# Séparation de l'ensemble de test par task

```python
def split_df_file_path(df):
    df = copy.deepcopy(df)
    df['task_name'] = df['file_path'].str.split('/').str[-2]
    return df
def filter_df_task_name(df,value):
    df = copy.deepcopy(df)
    df_filtered = df[df['task_name']==value]
    return df_filtered
def split_per_task(df_input,df_task,pickle_saving_path):
    df_split = split_df_file_path(df_input)
    list_tasks = df_split[df_split['split_value'] == 'TEST']['task_name'].unique()
    df_merged = pd.merge(df_split, df_task[['task_name','Site','Day','Day2']], on="task_name")
    for i in range(len(list_tasks)):
        df_for_pickle = filter_df_task_name(df_merged,list_tasks[i])
        pickle.dump(df_for_pickle, open( f"{pickle_saving_path}/tasks/{list_tasks[i]}.p", "wb" ) )
    pickle.dump(list_tasks, open( f"{pickle_saving_path}/list_tasks.p", "wb" ) )
```

```python
split_per_task(df_input,df_task,pickle_saving_path)
```
